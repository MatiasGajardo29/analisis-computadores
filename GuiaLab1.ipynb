```python
#PARTE1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np

df_computer = pd.read_csv("computer_prices_all.csv")

#primeras 5 filas
display(df_computer.head()) 
#dimensiones
print("Dimensiones:", df_computer.shape)
# Ver datos
df_computer.info()
```

```python
#PARTE3
print("datos faltantes:\n",df_computer.isna().sum()) #no hay datos nulos
print("Total de filas duplicadas:", df_computer.duplicated().sum()) # no hay datos duplicados
df_computer.describe()
```

```python
#PARTE4
plt.figure(figsize=(8, 5))
df_computer['price'].plot(kind='hist', bins=20,edgecolor='black')
plt.title('Distribución de Variable Numérica')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
# value_counts() cuenta cuántas veces aparece cada categoría
df_computer['brand'].value_counts().plot(kind='bar',
color='skyblue')
plt.title('Frecuencia por Marca')
plt.xlabel('Marca')
plt.ylabel('Cantidad (frecuencia)')
plt.xticks(rotation=45) # Rota el texto del eje X si es muy largo
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df_computer.plot(kind='scatter', x='ram_gb', y='price', alpha=0.6,
color='darkgreen')
plt.title('RAM vs Precio')
plt.xlabel('RAM (GB)')
plt.ylabel('Precio')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="device_type", y="price", data=df_computer)
plt.title("Precio por tipo de dispositivo")
plt.ylabel("Precio")
plt.xlabel("Tipos de dispositivos")
plt.xticks([0, 1], ["Laptop", "Escritorio"])
plt.show()
```
